### Techniques to determine order of AR(p) model:
    1. PACF - Measures incremental significance of adding another lag
    2. Information Criteria(BIC) - Prevents overfitting by imposing penalty
 We choose a model with the lowest information criteria

```python
import import_ipynb
import stationarity
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.ar_model import AutoReg
from statsmodels.graphics.tsaplots import plot_pacf
import warnings
warnings.filterwarnings('ignore')
```

```python
VZ = stationarity.ds["VZ"]
VZ.plot();
```

```python
plot_pacf(VZ)
plt.show()
```

```python
VZ_diff = stationarity.VZ_diff
VZ_diff.plot();
```

```python
# Fit the data to an AR(p) for p = 0,...,6
BIC = np.zeros(7)
for p in range(7):
    mod = AutoReg(VZ_diff, lags=p)
    res = mod.fit()  
    BIC[p] = res.bic
    
# Plot the BIC as a function of p
plt.plot(range(1,7), BIC[1:7], marker='o')
plt.title("AR models for Differenced VZ stock")
plt.xlabel('Order of AR Model')
plt.ylabel('Bayesian Information Criterion')
plt.show()
```

So, the order of AR model is 4 with lowest BIC
